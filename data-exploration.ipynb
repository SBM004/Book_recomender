```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books= pd.read_csv(f"{path}/books.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(),1,0)
books["age_of_group"] = 2024 - books["published_year"]
```

```python
column_of_interest = ["num_pages","age_of_group", "missing_description","average_rating"]

correlation_matrix = books[column_of_interest].corr(method = "spearman")
sns.set_theme(style="whitegrid")
plt.figure(figsize = (8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm" , cbar_kws={"label": "Correlation"})

heatmap.set_title("correlation heatmap")
```

```python
books[
    (books["description"].isna() ) |
    (books["num_pages"].isna() ) |
    (books["average_rating"].isna() ) |
    (books["published_year"].isna() )
]
```

```python
books_without_missing= books[
    ~(books["description"].isna()) &
    ~(books["num_pages"].isna()) &
    ~(books["average_rating"].isna())&
    ~(books["published_year"].isna())
]
```

```python
books_without_missing
```

```python
books_without_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
books_without_missing["words_in_description"]=books_without_missing["description"].str.split().str.len()
```

```python
books_25_word=books_without_missing[books_without_missing["words_in_description"]>=25]
```

```python
books_25_word
```

```python
books_25_word["title_and_subtitle"]=np.where(~books_25_word["subtitle"].isna(),books_25_word[["title","subtitle"]].astype(str).agg(": ".join, axis=1),books_25_word["title"])
```

```python
books_25_word
```

```python
books_25_word["tagged_description"] = books_25_word[["isbn13","description"]].astype(str).agg(" ".join, axis=1)
```

```python
(
    books_25_word
    .drop(["subtitle","title","age_of_group","words_in_description"], axis=1)
    .to_csv("books_cleaned.csv",index=False)
)
```
